# [미션] 영어 단어장(단어 관리) 프로그램

```python
words = {}  # 단어 목록 (단어: 정보 딕셔너리)

def w_add():
    # 단어 등록 기능
    word = input("새 단어: ")
    meaning = input(f"{word}의 뜻: ")

    if word in words:
        print(f"이미 존재하는 단어입니다: {word}")
        return

    words[word] = {
        "meaning": meaning,
    }

    print(f"단어 '{word}'가 등록되었습니다.")


def w_search():
    # 단어 검색 기능
    word = input("검색할 단어 입력: ")

    if word in words:
        print(f"\n단어: {word}")
        print(f"뜻: {words[word]['meaning']}")
    else:
        print(f"검색어 '{word}'를 찾을 수 없습니다.")


def w_delete():
    # 단어 삭제 기능
    word = input("삭제할 단어 입력: ")

    if word in words:
        del words[word]
        print(f"단어 '{word}'가 삭제되었습니다.")
    else:
        print(f"삭제할 단어 '{word}'를 찾을 수 없습니다.")


def w_show():
    # 모든 단어 보기 기능
    if not words:
        print("등록된 단어가 없습니다.")
        return

    for word, data in words.items():
        print(f"\n단어: {word}")
        print(f"뜻: {data['meaning']}")


while True:
    print("\n--- 선택메뉴 ---")
    print("1. 단어 등록")
    print("2. 단어 검색")
    print("3. 단어 삭제")
    print("4. 모든 단어 보기")
    print("0. 종료하기")

    choice = input("-----------------------선택: ")

    if choice == '1':
        w_add()
    elif choice == '2':
        w_search()
    elif choice == '3':
        w_delete()
    elif choice == '4':
        w_show()
    elif choice == '0':
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")
```
