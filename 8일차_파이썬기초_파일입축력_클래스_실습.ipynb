# 파일입축력 & 클래스 실습

## 파일 입출력 문제

### [실습문제] 회원 목록 읽어와서 팀으로 배정하기

```python
# 2.리스트 섞기
def list_shuffle(members):
    import random
    for i in range(5):
        random.shuffle(members)
        # print(f'리스트를 {i+1}번 섞었습니다.')
    print(f"섞은 후: {members}")
    return members

def member_split(members):
    # 3.팀별로 배정하기
    teams_dic, tmp = dict(), list() # 팀:회원 저장 딕셔너리, 임시 저장 리스트
    i, member_cnt = 1, 4
    for idx, member in enumerate(members):
        if idx>0 and idx%member_cnt==0:
            teams_dic[f'{i}팀'] = tmp
            print(f"{i:2>}팀 : {tmp}")      # 화면 출력
            fw.write(f"{i:2>}팀 : {tmp}\n")   # 파일 출력
            i += 1
            tmp = []
        tmp.append(member)

    if tmp:
        teams_dic[f'{i+1}팀'] = tmp
        print(f"{i}팀 : {tmp}")
        fw.write(f"{i}팀 : {tmp}")

    return teams_dic

# 1.파일 열고 회원 목록 읽기
f  = open("./members.txt", "r")  # 파일 열기 객체
fw = open("./teams.txt", "w")    # 파일 열기 객체
members = [ member.strip() for member in f.readlines() ]
print(f"전체 인원: {len(members)}")
print(f"섞기 전: {members}")

members = list_shuffle(members)
teams_dic = member_split(members)
f.close()   # 파일 객체 닫기
fw.close()  # 파일 객체 닫기
```

```python
#팀 만들어 수강생 나누기
import random

rfile = './members.txt'
wfile = './teams.txt'
tname= [f'{i+1:2}팀' for i in range(10)]
teams = dict()
membercnt = 4


def makeTeamMembers():
    # 파일열고 멤버 목록 읽기
    with open(rfile, 'r', encoding='utf-8') as f:
        members = [ member.replace('\n','') for member in f.readlines()]

    members = ' '.join(members).split()   #리스트에서 공백 제거하기
    random.shuffle(members)               # 리스트 섞기
    print(f'전체 인원수: {len(members)}')
    # print(members)

    # 팀당 membercnt 수 만큼 배정하기
    with open(wfile, 'w', encoding='utf-8') as f:
        num, key = 0, 0
        tmp = []
        while num 0 and num % membercnt==0:
                teams[tname[key]] = tmp
                f.write(f"{tname[key]} : {' '.join(tmp)}\n")   #파일에 쓰기
                tmp = []
                key += 1
            tmp.append(members[num])
            # print(f'num: {num} key: {key} tmp: {tmp} ')
            num += 1
        if tmp:
            teams[tname[key]] = tmp
            f.write(f"{tname[key]} : {' '.join(tmp)}\n")   #파일에 쓰기

    return teams

teams = makeTeamMembers()
print(f'전체 팀 수: {len(teams)}')
print(teams)
```
